```python
import pandas as pd

# Replace with your actual file path
df = pd.read_csv('/content/Titanic-Dataset.csv')

# Preview the data
df.head()
```

```python
# General info
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

```python
#visualization using various graphs and charts
#Histogram
import matplotlib.pyplot as plt
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Example for a single feature
sns.boxplot(data=df, x='Age')
plt.show()
```

```python
# Select only numeric columns for correlation other wise it will be showing value error
numeric_df = df.select_dtypes(include=['number'])

# Now create the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Numeric Features Only)")
plt.show()
```

```python
# Age vs Survival
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', kde=True)
plt.title('Age Distribution by Survival')
plt.show()

# Fare vs Survival
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare vs Survival')
plt.show()

# Survival rate by Pclass
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title('Survival Rate by Passenger Class')
plt.show()

# Survival by Sex
sns.barplot(data=df, x='Sex', y='Survived')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
print("Observations:")
print("- Survival rates were higher among 1st class passengers and females.")
print("- Higher fares are associated with a better chance of survival.")
print("- Younger passengers and children seem to have higher survival.")
print("- Most passengers traveled alone (SibSp and Parch = 0).")
```
